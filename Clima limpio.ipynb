```python
df_complete = pd.read_csv("/Users/mak/Desktop/Python challenge/df_complete_186_files.csv")
df_clima_filtered = pd.read_csv("/Users/mak/Desktop/Python challenge/Clima_limpio.csv")

# Parse dates (day-level)
df_complete["operating_date"] = pd.to_datetime(df_complete["operating_date"], errors="coerce").dt.normalize()
df_clima_filtered["date"] = pd.to_datetime(df_clima_filtered["date"], errors="coerce").dt.normalize()
df_clima_filtered["tavg"] = pd.to_numeric(df_clima_filtered["tavg"], errors="coerce")

# Optional: if clima has duplicate dates, keep one row per day
df_clima_filtered = df_clima_filtered.drop_duplicates(subset=["date"])

# Merge: preserves df_complete shape
df_final = df_complete.merge(
    df_clima_filtered[["date", "tavg"]],
    left_on="operating_date",
    right_on="date",
    how="left",
    validate="m:1"
).drop(columns=["date"])

# Checks
print("df_complete shape:", df_complete.shape)
print("df_final shape:   ", df_final.shape)
print("Missing tavg rows:", df_final["tavg"].isna().sum())

# If every day MUST match, enforce:
missing = df_final[df_final["tavg"].isna()]["operating_date"].drop_duplicates()
print("Unmatched days:", len(missing))
```

```python
output_csv = "/Users/mak/Desktop/Python challenge/complete data with weather.csv"
df_final.to_csv(output_csv, index=False, encoding="utf-8")
print("Saved:", output_csv)
```
