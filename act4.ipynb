```python
import string

def count_word_frequency(text):
    # Convert to lowercase (case insensitive)
    text = text.lower()


    # Split into words
    words = text.split()

    # Count frequencies using a dictionary
    freq = {}

    for word in words:
        if word in freq:
            freq[word] += 1
        else:
            freq[word] = 1

    # Keep only words that appear more than once
    result = {}

    for word in freq:
        if freq[word] > 1:
            result[word] = freq[word]

    return result
```

```python
count_word_frequency("banana apple banana")
```

```python
def combine_dictionaries(dict1, dict2):
    result = {}

    # 1.add all values from dict1
    for key in dict1:
        result[key] = dict1[key]

    # 2. add values from dict2
    for key in dict2:
        if key in result:
            result[key] += dict2[key]  # 3.add if common key ( se repite , se suma )
        else:
            result[key] = dict2[key]   # 4.keep original if new key ( no se repite, se mantiene )

    return result
```

```python
a = {"b": 1, "a": 1, "t": 1, "m": 1, "a":1, "n":1}
b = {"s":1, "u":1, "p":1, "e":1, "r":1, "m":1, "a":1, "n":1} 

combined = combine_dictionaries(a, b)

print(combined)
```
