```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick

# ---------- Base ----------
df = df1.copy()
df["operating_date"] = pd.to_datetime(df["operating_date"], errors="coerce").dt.normalize()
df["tavg"] = pd.to_numeric(df["tavg"], errors="coerce")
df["total_ticket"] = pd.to_numeric(df["total_ticket"], errors="coerce")

# Optional: keep real sales lines only
df = df[
    (df["is_modifier"] == False) &
    (df["action"] == "Venta")
].copy()

stores = sorted(df["sucursal"].dropna().unique())

# ---------- Per sucursal ----------
for s in stores:
    d = df[df["sucursal"] == s].copy()

    daily = (
        d.dropna(subset=["operating_date", "tavg", "total_ticket"])
         .groupby("operating_date", as_index=False)
         .agg(
             tavg=("tavg", "mean"),
             daily_ticket_sum=("total_ticket", "count")  # use "sum" if you want money
         )
    )

    daily["temp_group"] = np.where(daily["tavg"] >= 25, "warm", "cold")

    cold = daily.loc[daily["temp_group"] == "cold", "daily_ticket_sum"].values
    warm = daily.loc[daily["temp_group"] == "warm", "daily_ticket_sum"].values

    if len(cold) < 5 or len(warm) < 5:
        print(f"\n=== {s} ===")
        print("Not enough cold/warm days for stable comparison.")
        continue

    # 1) Boxplot
    plt.figure(figsize=(8, 5))
    plt.boxplot([cold, warm], tick_labels=["cold", "warm"], showmeans=True)
    plt.title(f"{s} | Daily Total Tickets: Cold vs Warm")
    plt.ylabel("Daily total_ticket (count)")
    plt.gca().yaxis.set_major_formatter(mtick.StrMethodFormatter("{x:,.0f}"))
    plt.grid(axis="y", alpha=0.2)
    plt.show()

    # 2) Balanced bootstrap
    n = min(len(cold), len(warm))
    rng = np.random.default_rng(42)
    iters = 5000
    diff = np.empty(iters)

    for i in range(iters):
        c = rng.choice(cold, size=n, replace=True)
        w = rng.choice(warm, size=n, replace=True)
        diff[i] = w.mean() - c.mean()  # warm - cold

    ci_low, ci_high = np.percentile(diff, [2.5, 97.5])

    print(f"\n=== {s} ===")
    print(f"Cold days: {len(cold)} | Warm days: {len(warm)}")
    print(f"Balanced mean diff (warm - cold): {diff.mean():,.2f}")
    print(f"95% CI: [{ci_low:,.2f}, {ci_high:,.2f}]")

    plt.figure(figsize=(8, 4.5))
    plt.hist(diff, bins=40, color="steelblue", alpha=0.8)
    plt.axvline(0, color="red", linestyle="--")
    plt.title(f"{s} | Balanced Bootstrap: Mean Diff (Warm - Cold)")
    plt.xlabel("Difference in daily total_ticket")
    plt.ylabel("Frequency")
    plt.grid(alpha=0.2)
    plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick


# --- 1) Build daily dataset ---
df = df1.copy()

df["operating_date"] = pd.to_datetime(df["operating_date"], errors="coerce").dt.normalize()
df["tavg"] = pd.to_numeric(df["tavg"], errors="coerce")
df["total_ticket"] = pd.to_numeric(df["total_ticket"], errors="coerce")

daily = (
    df.dropna(subset=["operating_date", "tavg", "total_ticket"])
      .groupby("operating_date", as_index=False)
      .agg(
          tavg=("tavg", "mean"),
          daily_ticket_sum=("total_ticket", "count")   # sum of tickets each day
      )
)

daily["temp_group"] = np.where(daily["tavg"] >= 25, "warm", "cold")

# --- 2) Fair visual comparison: distribution per day (not totals) ---
cold = daily.loc[daily["temp_group"] == "cold", "daily_ticket_sum"].values
warm = daily.loc[daily["temp_group"] == "warm", "daily_ticket_sum"].values



plt.figure(figsize=(8,5))
plt.boxplot([cold, warm], labels=["cold", "warm"], showmeans=True)
plt.title("Daily Total Tickets: Cold vs Warm")
plt.ylabel("Daily total_ticket (sum)")
plt.gca().yaxis.set_major_formatter(mtick.StrMethodFormatter('{x:,.0f}'))  # 700,000 style
plt.grid(axis="y", alpha=0.2)
plt.show()

# --- 3) Balanced bootstrap (same sample size from each group) ---
n = min(len(cold), len(warm))
rng = np.random.default_rng(42)
iters = 5000
diff = np.empty(iters)

for i in range(iters):
    c = rng.choice(cold, size=n, replace=True)
    w = rng.choice(warm, size=n, replace=True)
    diff[i] = w.mean() - c.mean()   # warm - cold

ci_low, ci_high = np.percentile(diff, [2.5, 97.5])

print(f"Cold days: {len(cold)} | Warm days: {len(warm)}")
print(f"Balanced mean diff (warm - cold): {diff.mean():,.2f}")
print(f"95% CI: [{ci_low:,.2f}, {ci_high:,.2f}]")

plt.figure(figsize=(8, 5))
plt.hist(diff, bins=40, color="steelblue", alpha=0.8)
plt.axvline(0, color="red", linestyle="--")
plt.title("Balanced Bootstrap: Mean Difference (Warm - Cold)")
plt.xlabel("Difference in daily total_ticket")
plt.ylabel("Frequency")
plt.grid(alpha=0.2)
plt.show()

```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick

# -----------------------------
# 0) Prep (ticket-level + daily)
# -----------------------------
df = df1.copy()

# Types
df["operating_date"] = pd.to_datetime(df["operating_date"], errors="coerce").dt.normalize()
df["captured_time"] = pd.to_datetime(df["captured_time"], errors="coerce")
df["tavg"] = pd.to_numeric(df["tavg"], errors="coerce")
df["total_ticket"] = pd.to_numeric(df["total_ticket"], errors="coerce")

# One weather value per day
day_weather = (
    df.dropna(subset=["operating_date", "tavg"])
      .groupby("operating_date", as_index=False)["tavg"]
      .mean()
)

# Build ticket-level table (avoid item-line duplication)
ticket_base = (
    df.sort_values("captured_time")
      .drop_duplicates("pdv_txn_id", keep="first")
      [["pdv_txn_id", "operating_date", "sucursal", "total_ticket"]]
)

# Flag tickets with cancel/comp actions (if any line in that ticket has such action)
adj_flag = (
    df.assign(
        adj_flag=df["action"].astype(str).str.contains("Cancel|Anulaci|Cortes", case=False, na=False).astype(int)
    )
    .groupby("pdv_txn_id", as_index=False)["adj_flag"]
    .max()
)

ticket = ticket_base.merge(adj_flag, on="pdv_txn_id", how="left")
ticket["adj_flag"] = ticket["adj_flag"].fillna(0).astype(int)

# Daily metrics
daily = (
    ticket.dropna(subset=["operating_date"])
          .groupby("operating_date", as_index=False)
          .agg(
              ticket_count=("pdv_txn_id", "count"),
              gross_sales=("total_ticket", "sum"),
              avg_ticket=("total_ticket", "mean")
          )
          .merge(day_weather, on="operating_date", how="left")
          .sort_values("operating_date")
)

# -----------------------------
# 1) Daily demand vs temperature (7-day rolling, dual axis)
# -----------------------------
daily["tickets_7d"] = daily["ticket_count"].rolling(7, min_periods=1).mean()
daily["tavg_7d"] = daily["tavg"].rolling(7, min_periods=1).mean()

fig, ax1 = plt.subplots(figsize=(12, 5))
ax1.plot(daily["operating_date"], daily["tickets_7d"], color="tab:blue", label="Tickets (7d)")
ax1.set_ylabel("Tickets per day (7d avg)", color="tab:blue")
ax1.tick_params(axis="y", labelcolor="tab:blue")
ax1.yaxis.set_major_formatter(mtick.StrMethodFormatter("{x:,.0f}"))

ax2 = ax1.twinx()
ax2.plot(daily["operating_date"], daily["tavg_7d"], color="tab:red", label="TAVG (7d)")
ax2.set_ylabel("TAVG (C)", color="tab:red")
ax2.tick_params(axis="y", labelcolor="tab:red")

ax1.set_title("Daily Tickets vs Temperature (7-day rolling)")
ax1.set_xlabel("Date")
ax1.grid(alpha=0.2)
plt.tight_layout()
plt.show()

# -----------------------------
# 2) Boxplot: daily tickets by 5C temperature bands
# -----------------------------
min_b = np.floor(daily["tavg"].min() / 5) * 5
max_b = np.ceil(daily["tavg"].max() / 5) * 5 + 5
bins = np.arange(min_b, max_b, 5)

daily["temp_band"] = pd.cut(daily["tavg"], bins=bins, right=False, include_lowest=True)
bands = [b for b in daily["temp_band"].cat.categories if (daily["temp_band"] == b).any()]
box_data = [daily.loc[daily["temp_band"] == b, "ticket_count"].values for b in bands]
labels = [f"{int(b.left)}-{int(b.right)}C" for b in bands]

plt.figure(figsize=(11, 5))
plt.boxplot(box_data, labels=labels, showmeans=True)
plt.title("Daily Ticket Count by Temperature Band (5C)")
plt.xlabel("Temperature band")
plt.ylabel("Ticket count per day")
plt.gca().yaxis.set_major_formatter(mtick.StrMethodFormatter("{x:,.0f}"))
plt.grid(axis="y", alpha=0.2)
plt.tight_layout()
plt.show()

# -----------------------------
# 3) Cancel/comp rate by store
# -----------------------------
store_adj = (
    ticket.groupby("sucursal", as_index=False)
          .agg(tickets=("pdv_txn_id", "count"), adjusted_tickets=("adj_flag", "sum"))
)
store_adj["adjust_rate_pct"] = 100 * store_adj["adjusted_tickets"] / store_adj["tickets"]
store_adj = store_adj.sort_values("adjust_rate_pct", ascending=False)

plt.figure(figsize=(10, 5))
plt.bar(store_adj["sucursal"], store_adj["adjust_rate_pct"], color="darkorange")
plt.title("Cancel/Comp Ticket Rate by Store")
plt.xlabel("Store")
plt.ylabel("Adjusted tickets (%)")
plt.xticks(rotation=45, ha="right")
plt.grid(axis="y", alpha=0.2)
plt.tight_layout()
plt.show()

# -----------------------------
# 4) Month x weekday heatmap (avg daily tickets)
# -----------------------------
daily["month"] = daily["operating_date"].dt.month
daily["weekday_num"] = daily["operating_date"].dt.weekday  # Mon=0 ... Sun=6

heat = (
    daily.pivot_table(index="month", columns="weekday_num", values="ticket_count", aggfunc="mean")
         .reindex(index=range(1, 13), columns=range(7))
)

fig, ax = plt.subplots(figsize=(10, 6))
im = ax.imshow(heat.values, aspect="auto", cmap="plasma")
ax.set_title("Average Daily Tickets: Month x Weekday")
ax.set_xlabel("Weekday")
ax.set_ylabel("Month")
ax.set_xticks(range(7))
ax.set_xticklabels(["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"])
ax.set_yticks(range(12))
ax.set_yticklabels(range(1, 13))
cbar = plt.colorbar(im, ax=ax)
cbar.set_label("Avg tickets/day")
plt.tight_layout()
plt.show()

# -----------------------------
# 5) Temperature sensitivity by store (slope tickets per 1C)
# -----------------------------
store_day = (
    ticket.groupby(["sucursal", "operating_date"], as_index=False)
          .agg(ticket_count=("pdv_txn_id", "count"))
          .merge(day_weather, on="operating_date", how="left")
          .dropna(subset=["tavg"])
)

slopes = []
for s, g in store_day.groupby("sucursal"):
    if len(g) >= 30 and g["tavg"].nunique() > 1:
        m, b = np.polyfit(g["tavg"], g["ticket_count"], 1)
        slopes.append((s, m, len(g)))

store_slope = pd.DataFrame(slopes, columns=["sucursal", "slope_tickets_per_c", "days"]).sort_values("slope_tickets_per_c")

plt.figure(figsize=(9, 5))
plt.barh(store_slope["sucursal"], store_slope["slope_tickets_per_c"], color="slateblue")
plt.axvline(0, color="black", linewidth=1)
plt.title("Store Temperature Sensitivity")
plt.xlabel("Change in tickets/day for +1C")
plt.ylabel("Store")
plt.grid(axis="x", alpha=0.2)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd

# 1) Build daily sales per sucursal (not row-level items)
base = df1[
    (df1["is_modifier"] == False) &
    (df1["action"] == "Venta")
].copy()

base["operating_date"] = pd.to_datetime(base["operating_date"], errors="coerce").dt.normalize()
base["total_ticket"] = pd.to_numeric(base["total_ticket"], errors="coerce")
base["tavg"] = pd.to_numeric(base["tavg"], errors="coerce")

daily = (
    base.sort_values("captured_time")
        .drop_duplicates("pdv_txn_id")
        .groupby(["sucursal", "operating_date"], as_index=False)
        .agg(
            sales=("total_ticket", "sum"),
            tavg=("tavg", "mean"),
            holiday_type=("holiday_type", "first")
        )
)


daily["holiday_group"] = daily["holiday_type"].fillna("No holiday")
daily["weekday"] = daily["operating_date"].dt.weekday
daily["month"] = daily["operating_date"].dt.month
daily["temp_bin"] = pd.cut(daily["tavg"], bins=[-100,10,15,20,25,30,100], labels=["<10","10-15","15-20","20-25","25-30","30+"])

# 2) Strata for matching
strata_cols = ["sucursal", "weekday", "month", "temp_bin"]

def matched_bootstrap_diff(df, group_a, group_b, iters=4000, seed=42):
    rng = np.random.default_rng(seed)
    diffs = []

    # use only strata where both groups exist
    g = df[df["holiday_group"].isin([group_a, group_b])].copy()
    by_strata = g.groupby(strata_cols, dropna=False)

    valid = []
    for key, s in by_strata:
        a = s.loc[s["holiday_group"] == group_a, "sales"].values
        b = s.loc[s["holiday_group"] == group_b, "sales"].values
        n = min(len(a), len(b))
        if n > 0:
            valid.append((a, b, n))

    if not valid:
        return None

    for _ in range(iters):
        a_all, b_all = [], []
        for a, b, n in valid:
            a_all.append(rng.choice(a, size=n, replace=True))
            b_all.append(rng.choice(b, size=n, replace=True))
        a_all = np.concatenate(a_all)
        b_all = np.concatenate(b_all)
        diffs.append(a_all.mean() - b_all.mean())  # A - B

    diffs = np.array(diffs)
    return {
        "mean_diff": diffs.mean(),
        "ci_low": np.percentile(diffs, 2.5),
        "ci_high": np.percentile(diffs, 97.5),
        "n_strata": len(valid)
    }

# 3) Compare separately vs non-holiday
res_puente = matched_bootstrap_diff(daily, "Puente", "No holiday")
res_festivo = matched_bootstrap_diff(daily, "Festivo Oficial", "No holiday")

print("Puente vs No holiday:", res_puente)
print("Festivo Oficial vs No holiday:", res_festivo)
```
