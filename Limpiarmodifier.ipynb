```python
columnas_datetime = ["operating_date", "closing_time", "captured_time"]
for column in columnas_datetime:
    if column in df1.columns:
        df1[column] = pd.to_datetime(df1[column], errors='coerce')
```

```python
df1["cold_or_warm"] = df1["tavg"].apply(
    lambda x: "warm" if x >= 25 else "cold"
)
```

```python
h = df1["captured_time"].dt.hour

# You can adjust ranges if needed:
# morning = 06:00-11:59
# between = 12:00-18:59
# night = 19:00-05:59
df1["day_part"] = np.select(
    [
        h.between(6, 11, inclusive="both"),
        h.between(12, 18, inclusive="both")
    ],
    [
        "morning",
        "afternoon"
    ],
    default="night")
```
