```python
def mystery(nums):
    nums = nums.copy() # Create a copy of the input list to avoid modifying the original
    for i in range(len(nums)):
        if i > 0:
            print(nums[i])
            nums[i] += nums[i - 1]
            print(nums[i])
    return nums
```

```python
mystery([1, 2, 3, 4])
```
