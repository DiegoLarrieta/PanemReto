```python
#dia y hora pico de ventas



df = df_complete_186_files.copy()
df["captured_time"] = pd.to_datetime(df["captured_time"], errors="coerce")
df = df.dropna(subset=["captured_time"])

df["month"] = df["captured_time"].dt.to_period("M")
df["date"] = df["captured_time"].dt.date
df["day_name"] = df["captured_time"].dt.day_name()
df["hour"] = df["captured_time"].dt.hour

# --- Day with highest MEAN sales in each month ---
# sales per date
daily_sales = (
    df.groupby(["month", "date"])
    .size()
    .reset_index(name="sales")
)
daily_sales["day_name"] = pd.to_datetime(daily_sales["date"]).dt.day_name()

# mean sales by day_name inside each month
day_mean = (
    daily_sales.groupby(["month", "day_name"])["sales"]
    .mean()
    .reset_index(name="mean_sales_day")
)

top_day_per_month = (
    day_mean.sort_values(["month", "mean_sales_day"], ascending=[True, False])
    .drop_duplicates("month")
)

# --- Hour with highest MEAN sales in each month ---
# sales per date-hour
hourly_sales = (
    df.groupby(["month", "date", "hour"])
    .size()
    .reset_index(name="sales")
)

# mean sales by hour inside each month
hour_mean = (
    hourly_sales.groupby(["month", "hour"])["sales"]
    .mean()
    .reset_index(name="mean_sales_hour")
)

top_hour_per_month = (
    hour_mean.sort_values(["month", "mean_sales_hour"], ascending=[True, False])
    .drop_duplicates("month")
)

# Final table
result = top_day_per_month.merge(top_hour_per_month, on="month", how="inner")
result = result[["month", "day_name", "mean_sales_day", "hour", "mean_sales_hour"]]
result = result.sort_values("month")

result
```

```python

ORDER_ID_COL = "order_id"
AMOUNT_COL = "total_ticket"          # total amount per row/item
DATE_COL = "captured_time"

df = df_complete_186_files.copy()
df[DATE_COL] = pd.to_datetime(df[DATE_COL], errors="coerce")
df = df.dropna(subset=[DATE_COL])

# Optional: keep only valid paid sales
# df = df[(df[AMOUNT_COL] > 0) & (df["status"] == "PAID")]

# Mean ticket overall (order total taken once per order)
ticket_per_order = df.groupby(ORDER_ID_COL)[AMOUNT_COL].first()
mean_ticket_overall = ticket_per_order.mean()

# Mean ticket by month
mean_ticket_month = (
    df.assign(month=df[DATE_COL].dt.to_period("M"))
      .groupby(["month", ORDER_ID_COL])[AMOUNT_COL].first()
      .reset_index()
      .groupby("month")[AMOUNT_COL].mean()
      .reset_index(name="mean_ticket")
      .sort_values("month")
)

mean_ticket_overall, mean_ticket_month.head(50)
```
