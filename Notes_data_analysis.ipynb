```python
data = [10, -5, 20, 0, 35, -2, 50, 105,30]

total = 0
count = 0

for n in data:
    if n < 0:
        continue      # 1) skip negatives
    if n > 100:
        break         # 2) stop if number > 100
    total += n
    count += 1

average = total / count if count else 0   # 3) compute average
print("The average is", average)

```

```python
#Voltear Palabras
def invertir_palabras(palabra):
    palabras = palabra.split()
    palabras_invertidas = [palabra[::-1] for palabra in palabras]
    return ' '.join(palabras_invertidas)

result = invertir_palabras("Hola mundo")
print(result)
```

```python
#PALINDROME
def isPalindrome1(sentence):
    joined_words = ''.join(sentence.lower().split())
    reversed_sentence = joined_words[::-1]
    
    return reversed_sentence == joined_words
    
```

```python
def filtrar_palabras(lista_palabras, n):
    return [palabra for palabra in lista_palabras if len(palabra) > n]
result = filtrar_palabras(['hola', 'adios', 'hola', 'adios', 'adios'], 3)
print(result)
```

```python
def clave_max_valor (dict):
    if not dict:
      return None
    else:
      return max(dict, key=dict.get)

resultado = clave_max_valor({"a": 1, "b": 2, "c": 3})
print(resultado)
```

```python
print(clave_max_valor({'a' : 10, 'b': 5, 'c': 8, 'd': 12}))  # Prints 'd'
```

Recursive function!

A recursive function is a function that calls itself AGAIN AND AGAIN.


```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)

print("Factorial of a number is:", factorial(8))
```
