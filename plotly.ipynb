```python
# ...existing code...
import glob, os
import pandas as pd

path = "/Users/diego/Documents/ChallengeAI/Complete Data"
csv_files = sorted(glob.glob(os.path.join(path, "*.csv")))
print(f"Found {len(csv_files)} CSV files in: {path}")

def safe_read(f):
    try:
        return pd.read_csv(f, low_memory=False)
    except Exception as e:
        print(f"Error reading {f}: {e}")
        return None

dfs = []
for f in csv_files:
    df = safe_read(f)
    if df is not None:
        df["_source_file"] = os.path.basename(f)
        dfs.append(df)

if not dfs:
    print("No CSVs loaded.")
else:
    # Concatenate (apilar filas). Útil cuando los CSV tienen las mismas columnas.
    df_all = pd.concat(dfs, ignore_index=True, sort=False)

    # Parsear columnas datetime comunes si existen
    for col in ["operating_date", "closing_time", "captured_time"]:
        if col in df_all.columns:
            df_all[col] = pd.to_datetime(df_all[col], errors="coerce")

    # Asignar al df1 usado en el notebook
    df1 = df_all
    print("df1.shape:", df1.shape)
    display(df1.head())

# Si en su lugar quieres hacer joins por una key (merge sucesivo), puedes usar:
def merge_on_key(files, key, how="outer", suffixes=("_x","_y")):
    dfsm = [pd.read_csv(f, low_memory=False) for f in files]
    out = dfsm[0]
    for d in dfsm[1:]:
        out = out.merge(d, on=key, how=how, suffixes=suffixes)
    return out
# ...existing code...
```

# Paso 1 — Limpiar datos importantes

### Primero aseguramos que no haya productos vacíos y que la cantidad sea válida.

```python
df_clean = df1.copy()

df_clean = df_clean.dropna(subset=["item", "quantity"])
df_clean["quantity"] = pd.to_numeric(df_clean["quantity"], errors="coerce")
df_clean = df_clean.dropna(subset=["quantity"])
```

# Paso 2 — Agrupar ventas por sucursal y producto

### Ahora calculamos cuánto se vende de cada producto en cada sucursal.

```python
sales = (
    df_clean.groupby(["sucursal", "item"])["quantity"]
    .sum()
    .reset_index()
)
```

```python
print("Filas en sales:", len(sales))
print("Sucursales:", sales["sucursal"].nunique())
print("Productos:", sales["item"].nunique())
```

# Paso 3 — Obtener el Top 5 productos por sucursal

### Esto es clave para el Sankey.

```python
top5_products = (
    sales.sort_values(["sucursal", "quantity"], ascending=[True, False])
    .groupby("sucursal")
    .head(5)
)
```

# Creacion del sankey plot 
### 1. crear nodos

```python
branches = top5_products["sucursal"].unique().tolist()
products = top5_products["item"].unique().tolist()

labels = branches + products
```

```python
label_index = {label: i for i, label in enumerate(labels)}
```

```python
source = top5_products["sucursal"].map(label_index).tolist()
target = top5_products["item"].map(label_index).tolist()
value = top5_products["quantity"].tolist()
```

```python
import plotly.graph_objects as go
```

```python
fig = go.Figure(data=[go.Sankey(
    node=dict(
        pad=15,
        thickness=20,
        label=labels
    ),
    link=dict(
        source=source,
        target=target,
        value=value
    )
)])

fig.update_layout(title_text="Panem Sales Flow: Branch → Top Products", font_size=12)

fig.show()
```
